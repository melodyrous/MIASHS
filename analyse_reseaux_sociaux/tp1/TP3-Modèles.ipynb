```python
from numpy.random import choice

def BA(n, m):
    def compute_proba(g):
        probas = []
        for degree in g.degree():
            probas.append(degree/ sum(g.degree()))
        return probas

    # initialisation du graphe
    start = 3
    g =  Graph()
    g.add_vertices(3)
    g.add_edges(([0,1],[0,2]))

    n = n
    m = m
    for i in range(start, n+start):
        probability_distribution = compute_proba(g)
        choices = choice(g.degree(), m,
                p=probability_distribution, replace=False)
        g.add_vertices(i)
        g.add_edges(([i,choices[0]], [i,choices[1]]))
    return g

g = BA(15, 2)
plot(g)
```

```python
# Graph Watts Strogatz

def make_treilli(n, m):
    g = Graph()
    g.add_vertices(np.arange(n))
    for i in range(n):
        for j in range(0, int(m/2)):
            g.add_edge(i, (i+j+1) % n)
    return g

g = make_treilli(10, 4)

layout = g.layout()
plot(g)
```
